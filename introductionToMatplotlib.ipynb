```python
x = [1, 2, 3, 4]
y = [11,22, 33, 55]
plt.plot(x,y)
plt.show()
```

```python
# 1st Method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd method
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # add some dat
plt.show()
```

```python
# 3rd method (recommended)
fig , ax = plt.subplot()
ax.plot(x, y); #add some data
```
