## Matplotlib example workflow


```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare  data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10)) # width, height

# 3. Plot data
ax.plot(x, y)

# 4.Customize  plot
ax.set(title="Simple plot",
        xlabel="x-axis",
        ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Hist
* Subplots

```python
# Plot the data and create a line plot
x = np.linspace(0, 10, 100)

x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use above data to create scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Create a plot from dictionary
nutButterPrices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12
                    }
fig, ax = plt.subplots()
ax.bar(nutButterPrices.keys(), nutButterPrices.values()); # (x, y)
ax.set(title="Butter store",
        ylabel="Prices ($)");
```
